```python
def load_pickled_data(data_path):
    """Reads a pickle file"""
    with open(data_path, "rb") as f:
        data = pickle.load(f)
    return data

def save_pickled_data(data, data_path):
    """Saves a pickle file"""
    with open(data_path, "wb") as f:
        data = pickle.dump(data,f)
    return data

def load_npy(filename):
    """Reads a npy file"""
    filename = f"{filename}.npy" if len(os.path.splitext(filename)[-1]) == 0\
        else filename
    with open(filename, "rb") as f:
        return np.load(f)

#https://stackoverflow.com/questions/12201577/how-can-i-convert-an-rgb-image-into-grayscale-in-python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

## Load Dataset

```python
class Transform_image(object):

    def __init__(self, img_height, img_width):
        self.img_width = img_width
        self.img_height = img_height

    def __call__(self, im):
        w, h = im.size
        if h > w:
            im = im.transpose(method=Image.ROTATE_270)
        
        im = im.resize((self.img_width, self.img_height))
        im = np.array(im)
        im = im.astype(np.float32)
        im = (im)/255.0
        return im
    

class Transform_mask(object):

    def __init__(self, img_height, img_width):
        self.img_width = img_width
        self.img_height = img_height

    def __call__(self, im):
        w, h = im.size

        if h > w:
            im = im.transpose(method=Image.ROTATE_270)

        im = im.resize((self.img_width, self.img_height), resample=0)
        im = np.array(im)/255
        im = (im[:, :, 0]).astype("int")
        return im

num_channels, img_height, img_width =  3, 384,512
image_transformer = Transform_image(img_height, img_width)
mask_transformer = Transform_mask(96, 128)
```

```python
class GenerateData:
    """
    Generate Data
    
    """
    train_img_paths = []
    val_img_paths = []
    mode = "train"
    
    def __init__(self, class_name=None, bs=32, train_test_split=0.75, shuffle=True):    
        self.bs = bs
        self.shuffle = shuffle
        # supply classname to generate for one class else data for all classes are generated
        self.class_name = class_name
        self.img_datasets = datasets.ImageFolder(root=os.path.join(data_root, "images"))
        self.transform_image = image_transformer
        self.transform_mask = mask_transformer
        for _class in classes:
            if self.class_name is not None:
                _class = self.class_name
            img_paths = [i[0] for i in self.img_datasets.imgs if _class in i[0]]
            len_train = int(len(img_paths) * train_test_split)
            self.train_img_paths += img_paths[:len_train]
            self.val_img_paths += img_paths[len_train:]
            if self.class_name is not None:
                break
                
        assert set(self.val_img_paths).intersection(set(self.train_img_paths)) == set(), "some imgs are both in train and val"
        self.train_mask_paths = [i.replace(".jpg", ".npy").replace("images", "masks") for i in self.train_img_paths]
        self.val_mask_paths = [i.replace(".jpg", ".npy").replace("images", "masks") for i in self.val_img_paths]
        self.train_images, self.train_masks = self._load_data(self.train_img_paths, self.train_mask_paths)
        self.val_images, self.val_masks = self._load_data(self.val_img_paths, self.val_mask_paths)
        self.train_classes = np.array([i.split(os.sep)[-2] for i in self.train_img_paths])
        self.val_classes = np.array([i.split(os.sep)[-2] for i in self.val_img_paths])
        self.train_class_counts = dict(Counter(self.train_classes))
        self.val_class_counts = dict(Counter(self.val_classes))
        self.train_indices = np.arange(len(self.train_images))
        
    
    def __getitem__(self, idx):
        img_paths =  self.img_paths[idx]
        mask_paths = self.mask_paths[idx]
        return self._load_data(img_paths, mask_paths)
    
    
    def _load_data(self, img_paths, mask_paths):
        img_paths  = [img_paths] if type(img_paths) == str else img_paths
        mask_paths  = [mask_paths] if type(mask_paths) == str else mask_paths           
        images = np.array([self.transform_image(Image.open(i)) for i in img_paths])
        masks = np.array([self.transform_mask(Image.fromarray(load_npy(i))) for i in mask_paths])
        return images, masks

    
    def __iter__(self):
        self.idx = 0
        if self.shuffle:
            np.random.shuffle(self.train_indices)
        return self
    
    def __next__(self):
        if self.mode == "train":
            imgs = self.train_images
            masks = self.train_masks
            if self.bs > len(imgs):
                self.bs = len(imgs)
                print("batch size is greater than max len", len(imgs))
                print("\n New batch size is ",self.bs)
            idxs = self.train_indices[self.idx*self.bs:(self.idx+1)*self.bs]
            imgs =  imgs[idxs]
            masks = masks[idxs]
            class_names = self.train_classes[idxs]
        else:
            imgs = self.val_images
            masks = self.val_masks
            if self.bs > len(imgs):
                self.bs = len(imgs)
                print("batch size is greater than max len", len(imgs))
                print("\n New batch size is ",self.bs)
            imgs =  imgs[self.idx*self.bs:(self.idx+1)*self.bs]
            masks = masks[self.idx*self.bs:(self.idx+1)*self.bs]
            class_names = self.val_classes[self.idx*self.bs:(self.idx+1)*self.bs]
            
        if len(imgs) == 0 or len(masks) == 0:
            raise StopIteration
        
        self.idx += 1
        return imgs, masks, class_names 
                  
    def length(self):
        if self.mode == "train":
            class_counts = self.train_class_counts
        else:
            class_counts = self.val_class_counts
        return class_counts
    
    def __len__(self):
        _len = np.sum([v for k, v in self.train_class_counts.items()] + 
                     [v for k, v in self.val_class_counts.items()])
        return _len
```

```python
#with tpu_strategy.scope():
base_model = tf.keras.applications.ResNet101(
    weights="imagenet",  # Load weights pre-trained on ImageNet.
    input_shape=(img_height, img_width, num_channels), #375,500
    include_top=False,
)  # Do not include the ImageNet classifier at the top.

layer_names = [
    'conv2_block1_out',   # 256 channels, stride 4
    'conv2_block2_out',   # 256 channels, stride 4
    'conv2_block3_out',   # 256 channels, stride 4
    'conv4_block23_out',  # 1024 channels, stride 16
]

layers = [base_model.get_layer(name).output for name in layer_names]
# Freeze the base_model
encoder = tf.keras.Model(inputs=base_model.input, outputs=layers)
encoder.trainable = False

inputs = tf.keras.Input(shape=((img_height, img_width, num_channels)))
# Downsampling through the model
skips = encoder(inputs, training=False)


conv1_1x1 = tf.keras.layers.Conv2D(filters=48, kernel_size=1, strides=1, padding='same', use_bias=False)
conv1a = tf.keras.layers.Conv2D(filters=256, kernel_size=3, strides=1, padding='same', dilation_rate=2, use_bias=False)
conv1a_bn = tf.keras.layers.BatchNormalization()
conv1b = tf.keras.layers.Conv2D(filters=256, kernel_size=3, strides=1, padding='same', dilation_rate=2, use_bias=False)
conv1b_bn = tf.keras.layers.BatchNormalization()

conv2_1x1 = tf.keras.layers.Conv2D(filters=48, kernel_size=1, strides=1, padding='same', use_bias=False)
conv2a = tf.keras.layers.Conv2D(filters=256, kernel_size=3, strides=1, padding='same', dilation_rate=4, use_bias=False)
conv2a_bn = tf.keras.layers.BatchNormalization()
conv2b = tf.keras.layers.Conv2D(filters=256, kernel_size=3, strides=1, padding='same', dilation_rate=4, use_bias=False)
conv2b_bn = tf.keras.layers.BatchNormalization()

conv3_1x1 = tf.keras.layers.Conv2D(filters=48, kernel_size=1, strides=1, padding='same', use_bias=False)
conv3a = tf.keras.layers.Conv2D(filters=256, kernel_size=3, strides=1, padding='same', dilation_rate=4, use_bias=False)
conv3a_bn = tf.keras.layers.BatchNormalization()
conv3b = tf.keras.layers.Conv2D(filters=256, kernel_size=3, strides=1, padding='same', dilation_rate=4, use_bias=False)
conv3b_bn = tf.keras.layers.BatchNormalization()

convfinal = tf.keras.layers.Conv2D(filters=2, kernel_size=1, strides=1, padding='same', use_bias=False)

upsample = tf.keras.layers.UpSampling2D(size=4, interpolation='bilinear')
relu1a = tf.keras.layers.Activation("relu")
relu1b = tf.keras.layers.Activation("relu")
relu2a = tf.keras.layers.Activation("relu")
relu2b = tf.keras.layers.Activation("relu")
relu3a = tf.keras.layers.Activation("relu")
relu3b = tf.keras.layers.Activation("relu")
concat = tf.keras.layers.Concatenate()

#upsample1 = tf.keras.layers.Conv2DTranspose(8, 3, strides=2,padding='same')
encoder_output = skips[-1]

print(encoder_output.shape)
#comments on input size are wrong
x = upsample(encoder_output)
skip1 = conv1_1x1(skips[-2])
x = concat([x, skip1])
x = conv1a(x)
x = conv1a_bn(x)
x = relu1a(x)
x = conv1b(x)
x = conv1b_bn(x)
x = relu1b(x)
skip2 = conv1_1x1(skips[-3])
x = concat([x, skip2])
x = conv2a(x)
x = conv2a_bn(x)
x = relu2a(x)
x = conv2b(x)
x = conv2b_bn(x)
x = relu2b(x)
skip3 = conv1_1x1(skips[-4])
x = concat([x, skip3])
x = conv3a(x)
x = conv3a_bn(x)
x = relu3a(x)
x = conv3b(x)
x = conv3b_bn(x)
x = relu3b(x)
output = convfinal(x)

model = tf.keras.Model(inputs=inputs, outputs=output)
model.summary()
```

```python
def compute_loss(model, x, masks):
    logits = model(x)
    scce = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
    scce_loss = scce(masks, logits)
    return scce_loss

def calc_iou_per_class(model, X, y, ious, class_name):
    for i in range(len(X)):
        logit = model(np.expand_dims(X[i], 0))
        target = np.expand_dims(y[i], 0).astype(int)  
        pred = np.argmax(logit.numpy(),axis=-1).astype(int)
        iou = np.sum(np.logical_and(target, pred))/np.sum(np.logical_or(target, pred))
        ious[class_name[i]] += iou
    return ious


def plot_prediction(model, imgs, masks):  
    fig = plt.figure(figsize=(30, 30))
    imgs = np.expand_dims(imgs, 0) if imgs.ndim==3 else imgs
    masks = np.expand_dims(masks, 0) if masks.ndim==2 else masks
    pred_masks = model(imgs).numpy()
    pred_masks = np.argmax(pred_masks, axis=-1)
    num_plots = len(imgs)
    for i in range(num_plots):
        fig.add_subplot(num_plots,3,i*3+1)
        plt.imshow((imgs[i]*255).astype("uint8"))

        fig.add_subplot(num_plots,3,i*3+2)
        plt.imshow(masks[i], cmap="gray")
        plt.title("ground truth")

        fig.add_subplot(num_plots,3,i*3+3)
        plt.imshow(pred_masks[i], cmap="gray")
        plt.title("predicted")
    plt.subplots_adjust(hspace=0.5)
    plt.show()
    return

def plot_stats(stats, col):
    fig = plt.figure(figsize=(10, 5))
    plt.plot(stats[col])
    plt.xlabel("epochs")
    plt.ylabel(col)
    plt.show()

@tf.function
def train_step(model, x, masks, optimizer):
    """Executes one training step and returns the loss.

    This function computes the loss and gradients, and uses the latter to
    update the model's parameters.
    """
    with tf.GradientTape() as tape:
        loss = compute_loss(model, x, masks)
    gradients = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))
    return loss
```

## Train

```python
tf.keras.backend.clear_session()
batch_size = 8
tr_val_split = 0.75
dataloader = GenerateData(bs=batch_size, train_test_split=tr_val_split)
num_samples = len(dataloader)
epochs = 200
freq = 20
bs = 20
lr = 1e-3
training_stats = []
epoch_times = []
optimizer = tf.keras.optimizers.Adam(lr)
train_ious_list = []
val_ious_list = []

for epoch in range(1, epochs + 1):
    start_time = time.time()
    train_batch_loss = []
    val_batch_loss = []
    dataloader.mode = "train"
    for imgs, masks, _ in dataloader:
        batch_loss = train_step(model, imgs, masks, optimizer)
        train_batch_loss.append(batch_loss.numpy())
    
    dataloader.mode = "val"
    for imgs, masks, _ in dataloader:
        batch_loss = compute_loss(model, imgs, masks)
        val_batch_loss.append(batch_loss.numpy())

    train_loss = float(np.mean(train_batch_loss))
    val_loss = float(np.mean(val_batch_loss))
        
    train_ious = defaultdict(int)
    val_ious = defaultdict(int)
    train_per_class = dataloader.train_class_counts
    val_per_class = dataloader.val_class_counts
    dataloader.mode = "train"
    for imgs, masks, class_names in dataloader:
        train_ious = calc_iou_per_class(model, imgs, masks, train_ious, class_names)
      
    dataloader.mode = "val"
    for imgs, masks, class_names in dataloader:
        val_ious = calc_iou_per_class(model, imgs, masks, val_ious, class_names)
    
    train_iou = {k:train_ious[k]/train_per_class[k] for k in train_ious}
    val_iou = {k:val_ious[k]/val_per_class[k] for k in val_ious}
    train_ious_list.append(train_iou)
    val_ious_list.append(val_iou)
    
    end_time = time.time()
    epoch_time = (end_time - start_time)/60
    
    if epoch % freq == 1:
        print(f"Epoch:{epoch:03d}- Epoch Time:{epoch_time:3f} train loss:{train_loss:3f}, val loss:{val_loss:3f}")
        print("train_iou")
        pp.pprint(train_iou)
        print("val_iou")
        pp.pprint(val_iou)
    epoch_times.append(epoch_time)
   
    training_stats.append({
      "epoch":epoch,
      "train loss":train_loss,
      "val loss":val_loss,
      "epoch time": epoch_time,
    })
    
    if epoch % freq == 1:
        plot_prediction(model, imgs, masks)
        plot_stats(pd.DataFrame(training_stats), "train loss")
        plot_stats(pd.DataFrame(training_stats), "val loss")
```
